## 2. Training
If you're familiar with CLI (Command-Line Interface), you can simply train the SevenNet! For training, we need a yaml file as an input. Let's make a simple input.yaml file

```python
!mkdir sevennet_tutorial # make a folder for training
%cd '{working_dir}/sevennet_tutorial'
```

```python
# make a input.yaml file
%%writefile input.yaml
model:
  chemical_species: auto
  cutoff: 5.0

train:
  device: cuda   # if you don't have gpu, write 'cpu'
  is_train_stress: False
  epoch: 100
  optim_param:
    lr: 0.005
  scheduler_param:
    gamma: 0.99

data:
  batch_size: 10
  load_dataset_path: '/content/BOTNet-datasets/dataset_3BPA/train_300K.xyz' # write a path of training data
```

Let's start first training! It takes about 2 minutes to initialize and 6 minutes to train 100 epochs.
You can reduce epoch to train quickly.

```python
!sevenn input.yaml -s
```

Best model is saved as 'checkpoint_best.pth'. You can also check the training log at log.sevenn

## 3. SevenNet usage example: ASE MD
You can run ASE MD using trained model.

```python
from ase import units
from ase.md.langevin import Langevin
from ase.io import read, write
import numpy as np
import time
import os
from sevenn.sevennet_calculator import SevenNetCalculator

working_dir = '/content'
best_model_path = os.path.join(working_dir, 'sevennet_tutorial/checkpoint_best.pth')
calculator = SevenNetCalculator(model=best_model_path, device='cuda') # write device=cpu if you don't have gpu

init_conf = read('/content/BOTNet-datasets/dataset_3BPA/test_300K.xyz', '0')
init_conf.set_calculator(calculator)

dyn = Langevin(init_conf, 0.5*units.fs, temperature_K=310, friction=5e-3)
def write_frame():
  dyn.atoms.write('md_3bpa.xyz', append=True)
dyn.attach(write_frame, interval=50)
dyn.run(100)
print("MD finished!")
```
